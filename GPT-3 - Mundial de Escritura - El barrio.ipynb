````python
def get_sentences(prompt, temperature=0.7, max_tokens=200,  stop=".", top_p=1, 
                  n=10, frequency_penalty=0, presence_penalty=0, 
                  do_postprocess=True,
                  do_print=True):
    
    response = openai.Completion.create(
                    engine="davinci",
                    prompt=prompt,
                    temperature=temperature,
                    max_tokens=max_tokens,
                    stop=stop,
                    top_p=top_p,
                    n=n,
                    frequency_penalty=frequency_penalty,
                    presence_penalty=presence_penalty
                    )
    sentences = [choice['text'] for choice in response['choices']]
    
    if do_print:
        print(prompt)
        for i, sentence in enumerate(sentences, 1):
            print("------------")
            print(f"{i} - {sentence}")
    
    if do_postprocess:
        return sentences
    else:
        return response
````

````python
def interactive_writing(texts, temperature=0.7, max_tokens=200,  stop=".", top_p=1, 
                        n=10, frequency_penalty=0, presence_penalty=0, add_dot=True):
    
    while True:
        response = openai.Completion.create(
                        engine="davinci",
                        prompt=texts[-1],
                        temperature=temperature,
                        max_tokens=max_tokens,
                        stop=stop,
                        top_p=top_p,
                        n=n,
                        frequency_penalty=frequency_penalty,
                        presence_penalty=presence_penalty
                        )
        sentences = [choice['text'] for choice in response['choices']]

        print(texts[-1])
        for i, sentence in enumerate(sentences):
            print("------------")
            print(f"{i} - {sentence}")
        print("------------")
        print(f"{i+1} - Salir")
        
        user_choice = int(input("Elija la opcion: ", ))
        if user_choice < n:
            texts.append(texts[-1] + sentences[user_choice] + "." if add_dot else "")
        else:
            print("Adios, humano")
            break
    return texts
    
    
````

Chacarita: mi barrio

Chacarita no tiene identidad. O mejor dicho, sí la tiene: es la tensión entre el tránsito obrero matinal de Lacroze y Corrientes y el tránsito snob nocturno de Serrano y Honduras. Las enormes arboladas proyectan su sombra sobre las calles de adoquines trayendo una paz que ya es difícil encontrar en  cualquier lugar. Aunque no lo parezca, Chacarita reúne todos los estereotipos de los barrios porteños. Así como vos podes encontrarlo todo en Palermo, también es posible encontrarlo en Chacarita.


Los vecinos lo dicen así: "Me dieron Chacarita y me quedé con Chacarita". Y así como los jóvenes del barrio dicen "Ya me voy pa Chaca" cuando se van de fiesta, también los vecinos dejan de lado sus conflictos y sus problemas cuando llega la hora de ir a la cancha.


Los tiempos en Chacarita son los tiempos del barrio. Son tiempos donde los trabajadores se retiran muy temprano, donde las tardes se pasan en los bares y donde los domingos se pasea con la familia. Y si se anda de fiesta por la noche, no tiene importancia: si toca el clásico de los Rolling Stones, se baila; si toca el "Cumbia Rock" de Calamaro, se baila. Y si toca el himno del barrio, se canta.


Porque Chacarita es un barrio como el que siempre quisiste vivir: es un barrio donde la gente va a la cancha y sueña con un gol de Román. Y es un barrio donde te podes encontrar con el cantante de vieja rola de los 80. Es un barrio que puede resistir a la modernidad y a la globalización. Porque Chacarita no es solamente un barrio, es una idea. Es una idea que se impone a todas las demás. Hay que vivirlo y sentirlo para entenderlo. Es un barrio que te enseña que hay que luchar por lo que uno cree. Y Chacarita es, al fin y al cabo, el símbolo de la lucha.


No es un accidente que Chacarita sea el barrio que tiene el club social más antiguo del país. El club Atlético Chacarita fue fundado en 1908 y es el club social más antiguo del país. Su origen es político: surgió del Club Social y Deportivo Victoria, que se creó en los años '90 para desafiar a los clubes ingleses del barrio y, por lo tanto, a la oligarquía porteña. Esto fue posible gracias a la actividad de los socialistas de talleres y de los anarquistas de la Villa del Parque. Estos grupos se unieron al Club Social y Deportivo Victoria (fundado en 1904) y lo transformaron en el club Atlético Chacarita Juniors, que tuvo su primer encuentro de fútbol en 1908, con un equipo formado por unos 25 jugadores. El primer partido fue contra el Club Atlético del Parque y ganó por 6 a 0.


Más tarde, en 1913, el club se convirtió en una entidad deportiva social, que tuvo un proceso de expansión y se transformó en un club multiesportivo, con grandes figuras que han sido campeonas nacionales y que han tenido una trayectoria exitosa en Europa. En la actualidad es un club diverso, que incluye a todos los barrios del sur de la ciudad y que tiene una fuerte identidad popular.


Pero como en todo barrio, Chacarita tiene su lado oscuro. Quizás lo más emblemático sea la historia de la represión contra los vecinos y el club, llevada a cabo por el gobierno de facto de Martínez de Hoz, que condujo a la intervención del club y a la cárcel a varios dirigentes, entre ellos Pancho Gavazzo y Miguel Brancato, quienes fueron ejecutados en 1975.Los gobiernos posteriores, que siguieron una política de reconciliación, recuperaron al club y llegaron a un acuerdo con los herederos de los familiares de los fusilados.

```python
# Download .bin from here:
# https://stackoverflow.com/questions/46433778/import-googlenews-vectors-negative300-bin
W2V_BIN = "/media/dataista/DATA/GoogleNews-vectors-negative300.bin"
```

```python
# !pip install gensim
from gensim.models import KeyedVectors
w2v = KeyedVectors.load_word2vec_format(W2V_BIN, binary=True)
```

```python
V = list(w2v.vocab.keys())
```
